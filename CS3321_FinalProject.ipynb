```python
import csv
def readCSV(file):
    fh = open(file,'r')                                  # open file for reading
    data = csv.reader(fh,delimiter=',',quotechar='"')    # create a csv reader.  this is needed because some titles
                                                         # contain commas
    contents = []                                        # create an empty list in which to store the data 
    count = 0
    for line in data:                                    # process each line of the file
        count += 1
        if(count==1):                                    # skip the first line which contains column headers
            continue
        lineData = tuple(line)                           # store the data as a tuple (for consistency with the previous example)
        contents.append(lineData)                     # add the tuple of data to the list.

    return contents

student_data = readCSV('StudentData.csv')
number = []
student = []
name = []
credit_hour = []
mid = []
final =[]
for line in student_data:
    number.append(line[0])
    credit_hour.append(line[4])
    mid.append(line[5])
    final.append(line[6])
    student.append(line[1])
    name.append (line[2])
index = []
student_ID = []
student_name = []
for i in range(len(number)):
    if number[i] != '':
        index.append(i)
        student_ID.append(student[i])
        student_name.append(name[i])
        

def gradelettercal(mid,final):
    gradelist= []
    for i in range(len(number)):
        grade = (float(mid[i])* (40/100) + float(final[i])* (60/100))
        gradeletter = ''
        if grade >= 90.0:
            gradeletter ='A'
            gradelist.append(gradeletter)
        elif grade >= 80.0 and grade < 90.0:
            gradeletter ='B'
            gradelist.append(gradeletter)
        elif grade >= 70.0 and grade < 80.0:
            gradeletter ='C'
            gradelist.append(gradeletter)
        elif grade >= 60 and grade <70:
            gradeletter ='D'
            gradelist.append(gradeletter)
        else: 
            gradeletter = 'F'
            gradelist.append(gradeletter)
    return gradelist   

def gradepointcal(gradelist,credits):
    gradepoint = []
    grade_points = 0
    for i,grade in enumerate(gradelist):
        if grade == 'A':
            grade_points = 4.0*float(credits[i])
            gradepoint.append(grade_points)
        elif grade == 'B':
            grade_points = 3.0*float(credits[i])
            gradepoint.append(grade_points)
        elif grade == 'C':
            grade_points = 2.0*float(credits[i])
            gradepoint.append(grade_points)
        elif grade == 'D':
            grade_points = 1.0*float(credits[i])
            gradepoint.append(grade_points)
        elif grade == 'E':
            grade_points = 0.0*float(credits[i])
            gradepoint.append(grade_points)
        elif grade == 'F':
            grade_points = 0.0*float(credits[i])
            gradepoint.append(grade_points)
    return gradepoint

grade_list = gradelettercal(mid,final)
points_list = gradepointcal(grade_list,credit_hour)
total_point_list =[]
total_credit_list =[]

for i,v in enumerate(index):
    if i == 0:
        continue     
    total_points = 0
    total_credits = 0
    for j in range(index[i-1],index[i]):
        total_points += points_list[j]
        total_credits += int(credit_hour[j]) 
    total_point_list.append(total_points)
    total_credit_list.append(total_credits)
for k in range(index[len(index)-1],len(number)-1):
    total_points += points_list[k]
    total_credits += int(credit_hour[k])
total_point_list.append(total_points)
total_credit_list.append(total_credits)

output_file= open('output.txt','w')
gpa_list = []
output_file.write( "Student ID " + 'Student Name'+',' + 'GPA'+'\n')
for i in range(len(index)):
    gpa = round(total_point_list[i]/total_credit_list[i],3)
    gpa_list.append(gpa)
    output_file.write(student_ID[i] + ',' )
    output_file.write (student_name[i]+',')
    output_file.write(str(gpa)+ '\n')
output_file.close()
```
